## 1 Від'ємний біноміальний розподіл

### 1.1 Щільність

Виведіть значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
from scipy.stats import nbinom
import numpy as np
import matplotlib.pyplot as plt
n = 5
p = 0.4
x = np.arange(-0.5, n*5+1, 0.5)
a = nbinom.pmf(x, n=5, p=0.4)
print(a)

```

Зобразіть графічно значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
from scipy.stats import nbinom
import numpy as np
import matplotlib.pyplot as plt
x1 = np.arange(0, n*5+1)
a = nbinom.pmf(x1, n=5, p=0.4)

plt.stem(x1, a)
plt.show()

```

### 1.2 Функція розподілу

Виведіть значення функції розподілу від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
y = nbinom.cdf(x, n, p)
print(y)

```

Зобразіть графічно значення функції розподілу від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
plt.plot(x, y, 'r', drawstyle='steps-post',label='nbinom cdf');
plt.legend()
plt.title('Cumulative distribution function of a negative binomial distribution')
plt.show()

```

### 3.1 Біноміальний розподіл

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин біноміального розподілу з параметрами:
1. $n = 15, p = 0.8$
2. $n = 15, p = 0.2$
3. $n = 20, p = 0.4$
4. $n = 20, p = 0.6$
5. $n = 7, p = 0.5$
6. $n = 25, p = 0.5$
7. $n = 100, p = 0.5$

```python
from scipy.stats import binom


p = [[15, 0.8], [15, 0.2], [20, 0.4], [20, 0.6], [7, 0.5], [25, 0.5], [100, 0.5]]
for n, i in p:
    x = np.arange(0, n+1) 
    density = binom.pmf(x, n, i)
    plt.subplot(1, 3, 1)
    plt.stem(x, density)
    
    
    distribution = binom.cdf(x, n, i)
    plt.subplot(1, 3, 2)
    plt.plot(x, distribution, drawstyle='steps-post', color = 'r')
    plt.title('For case when n = {}, p = {}'.format(n, i))
    
    
    gen = binom.rvs(n, i, size=10000)
    plt.subplot(1, 3, 3)
    plt.hist(gen, color = 'g')
    plt.show()

```

### 3.2 Геометричний розподіл

### 3.3 Розподіл Пуассона

### 3.4 Від'ємний біноміальний розподіл

1. $n = 15, p = 0.8$
2. $n = 15, p = 0.2$
3. $n = 20, p = 0.4$
4. $n = 20, p = 0.6$
5. $n = 7, p = 0.5$
6. $n = 25, p = 0.5$
7. $n = 100, p = 0.5$

```python
from scipy.stats import nbinom
import numpy as np
import matplotlib.pyplot as plt

def nbinpmf():
    fig = plt.figure(figsize=(20,5))
    fig.set_size_inches(17, 15)
    plt.subplots_adjust(hspace=0.4)
    t =  [(15,0.8), (15,0.2), (20,0.4), (20,0.6), (7,0.5), (25,0.5), (100,0.5)]
    color = ['r', 'b', 'g', 'y', 'c', 'm', 'k']
    for i in range(7):
        x = np.arange(0, t[i][0]*2+1)
        b = nbinom.pmf(x, t[i][0], t[i][1])
        plt.subplot(3, 3, i+1)
        plt.stem(x, b, color[i], label="n={}, p={}".format(t[i][0], t[i][1]))
        plt.legend()
        plt.title('Probability mass function')
        plt.legend()
    plt.show()
    
nbinpmf()
def nbincdf():
    fig = plt.figure(figsize=(20,5))
    fig.set_size_inches(17, 15)
    plt.subplots_adjust(hspace=0.4)
    t =  [(15,0.8), (15,0.2), (20,0.4), (20,0.6), (7,0.5), (25,0.5), (100,0.5)]
    color = ['r', 'b', 'g', 'y', 'c', 'm', 'k']
    for i in range(7):
        x = np.arange(0, t[i][0]*2+1)
        b = nbinom.cdf(x, t[i][0], t[i][1])
        plt.subplot(3, 3, i+1) #xlim=(0, 155), ylim=(0, 1.1)
        plt.plot(x, b, color[i], drawstyle='steps-post', label="n={}, p={}".format(t[i][0], t[i][1]))
        plt.legend()
        plt.title('Cumulative distribution function')
        plt.legend()
    plt.show()


nbincdf()
def nbinrvs():
    fig = plt.figure(figsize=(20,5))
    fig.set_size_inches(15, 13)
    plt.subplots_adjust(hspace=0.4)
    t =  [(15,0.8), (15,0.2), (20,0.4), (20,0.6), (7,0.5), (25,0.5), (100,0.5)]
    color = ['r', 'b', 'g', 'y', 'c', 'm', 'k']
    for i in range(7):
        x = np.arange(0, t[i][0]*2+1)
        b = nbinom.rvs(t[i][0], t[i][1], size=10000)
        plt.subplot(3, 3, i+1)
        plt.hist(b, color=color[i], label="n={}, p={}".format(t[i][0], t[i][1]))
        plt.xlabel('numbers')
        plt.ylabel('times')
        plt.legend()
        plt.title('10000 random variates')
        plt.legend()

    plt.show()
    
nbinrvs()
```

## 4 Апроксимація біноміального розподілу розподілом Пуассона

Покажіть схожість біноміального розподілу та розподілу Пуассона (їх функцій щільності) з відповідними параметрами.
Наприклад:
- $n = 100, p = 0.05$
- $n = 100, p = 0.1$
- $n = 100, p = 0.15$
- $n = 100, p = 0.2$
- $n = 100, p = 0.35$
- $n = 100, p = 0.5$
- $n = 100, p = 0.75$

```python
p = [[100,0.05],[100, 0.1], [100,0.15], [100,0.2], [100,0.35], [100,0.5], [100,0.75]]

for n,i in p:
    x = np.arange(0, n+1)
    density1 = binom.pmf(x, n, i)
    density2 = poisson.pmf(x, i*n) # multiply by n for better scale
    plt.stem(x, density1, 'r',   label='binom density') 
    plt.stem(x, density2, 'g',   label='poisson density')
    plt.legend()
    plt.show()

```
